```python
def calcular_precio(fruta, kilos):
    dicc = {
        'manzana': 2.5,
    'banana': 1.2,
    'naranja': 1.8,
    'pera': 3.0
    }

    if fruta.lower() in dicc.keys():
        return dicc[(fruta.lower())]* kilos
```

```python
def que_toca_hoy(dia):    

    gym_comida = {
        'Lunes':'piernas',
        'Martes':'brazos',
        'Miercoles':'abdomen',
        'Jueves':'cardio',
        'Viernes':'cerveza',
        'Sabado':'pizza',
        'Domingo':'descansa'
    }
    

    try:
        return gym_comida[dia]
    except KeyError:
        return 'Dia no valido'

que_toca_hoy('Viernes')
```
