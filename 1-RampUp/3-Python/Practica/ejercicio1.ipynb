```python
import random

tantos = ['A','2','3','4','5','7','8','9','10','J', 'Q', 'K','A']
palos = ['♣','♦','♠','♥']
baraja = []

for tanto in tantos:
    for palo in palos:
        baraja.append(f"{tanto}, {palo}")

random.sample(baraja,5)
```

```python
from random import sample
def generador_contraseñas(longitud):
    abc_minusculas = "abcdefghijklmnopqrstuvwxyz"
    abc_mayusculas = abc_minusculas.upper()

    numeros = "0123456789"
    simbolos = "{}[]()*;/,_-"

    secuencia = abc_minusculas + abc_mayusculas + numeros + simbolos

    contraseña = sample(secuencia, longitud)

    resultado = "".join(contraseña)

    return resultado

password = generador_contraseñas(10)

print("contraseña: ", password)
```

```python
def factorial(num:int) -> int:
    
    if num == 1:
        return 1
    
    else:
        return num * factorial(num-1)
    
factorial(5)
```
