### Procesamiento de texto para estructuración de datos

```python
def extraer_caracteristica(texto, clave):
    """Extrae una característica específica del texto."""
    partes = texto.lower().split(clave.lower())
    if len(partes) > 1:
        valor = " ".join(partes[1].split()[:2])  
        return valor.strip()
    return None

def extraer_todas_caracteristicas(descripcion):
    """Extrae todas las características de una descripción."""
    return {
        'cap-shape': extraer_caracteristica(descripcion, "forma de sombrero"),
        'cap-surface': extraer_caracteristica(descripcion, "superficie"),
        'cap-color': extraer_caracteristica(descripcion, "color del sombrero"),
        'bruises': extraer_caracteristica(descripcion, "moretones"),
        'odor': extraer_caracteristica(descripcion, "olor"),
        'gill-attachment': extraer_caracteristica(descripcion, "tipo de branquias"),
        'gill-spacing': extraer_caracteristica(descripcion, "separación de branquias"),
        'gill-size': extraer_caracteristica(descripcion, "tamaño de branquias"),
        'gill-color': extraer_caracteristica(descripcion, "color de branquias"),
        'stalk-shape': extraer_caracteristica(descripcion, "forma del tallo"),
        'stalk-root': extraer_caracteristica(descripcion, "raíz del tallo"),
        'stalk-surface-above-ring': extraer_caracteristica(descripcion, "superficie del tallo por encima del anillo"),
        'stalk-surface-below-ring': extraer_caracteristica(descripcion, "superficie del tallo por debajo del anillo"),
        'stalk-color-above-ring': extraer_caracteristica(descripcion, "color del tallo por encima del anillo"),
        'stalk-color-below-ring': extraer_caracteristica(descripcion, "color del tallo por debajo del anillo"),
        'veil-type': extraer_caracteristica(descripcion, "tipo de velo"),
        'veil-color': extraer_caracteristica(descripcion, "color del velo"),
        'ring-number': extraer_caracteristica(descripcion, "número de anillos"),
        'ring-type': extraer_caracteristica(descripcion, "tipo de anillo"),
        'spore-print-color': extraer_caracteristica(descripcion, "color de la impresión de esporas"),
        'population': extraer_caracteristica(descripcion, "población"),
        'habitat': extraer_caracteristica(descripcion, "hábitat")
    }
```

# Procesar todas las descripciones


```python
df_procesado = df["descripcion"].apply(extraer_todas_caracteristicas).apply(pd.Series)
```

# Convertir columnas float64 a string y reemplazar NaN con "?

# Convertir variables categóricas a numéricas

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
label_encoders = {}
for col in df_cleaned.columns:
    le = LabelEncoder()
    df_cleaned[col] = le.fit_transform(df_cleaned[col])
    label_encoders[col] = le
```

# Dividir datos

```python
X = df_cleaned.drop(columns=["odor"])
y = df_cleaned["odor"]
```

# Balanceo de clases con SMOTE

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

# Entrenar modelo

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred_valid = model.predict(X_valid)
accuracy = accuracy_score(y_valid, y_pred_valid)
print(f"Precisión del modelo en validación: {accuracy:.4f}")
```
